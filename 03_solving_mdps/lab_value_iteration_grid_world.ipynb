#### Enter all code here (you may also use multiple cells)

```python
import numpy as np

class GridWorld:
    def __init__(self, nrows, ncols):
        self.nrows = nrows
        self.ncols = ncols
        self.grid = np.zeros((nrows, ncols))

    def get_reward(self, row, col):
        # Terminal state with reward +1 in the bottom right corner
        if row == self.nrows - 1 and col == self.ncols - 1:
            return 1
        # All other states have reward 0
        return 0

    def value_iteration(self, theta=0.01, discount_factor=0.9):
        value_func_array = np.zeros((self.nrows, self.ncols))

        print("Initial Value Function:\n", value_func_array)

        iteration = 0
        while True:
            iteration += 1
            delta = 0
            new_value_func_array = np.copy(value_func_array)

            for row in range(self.nrows):
                for col in range(self.ncols):
                    if row == self.nrows - 1 and col == self.ncols - 1:
                        continue

                    action_values = []

                    if row > 0:
                        action_values.append(self.get_reward(row - 1, col) + discount_factor * value_func_array[row - 1, col])
                    else:
                        action_values.append(self.get_reward(row, col) + discount_factor * value_func_array[row, col])

                    if row < self.nrows - 1:
                        action_values.append(self.get_reward(row + 1, col) + discount_factor * value_func_array[row + 1, col])
                    else:
                        action_values.append(self.get_reward(row, col) + discount_factor * value_func_array[row, col])

                    if col > 0:
                        action_values.append(self.get_reward(row, col - 1) + discount_factor * value_func_array[row, col - 1])
                    else:
                        action_values.append(self.get_reward(row, col) + discount_factor * value_func_array[row, col])

                    if col < self.ncols - 1:
                        action_values.append(self.get_reward(row, col + 1) + discount_factor * value_func_array[row, col + 1])
                    else:
                        action_values.append(self.get_reward(row, col) + discount_factor * value_func_array[row, col])

                    if action_values:
                        best_action_value = max(action_values)
                        new_value_func_array[row, col] = best_action_value
                    else:
                        pass

                    delta = max(delta, abs(new_value_func_array[row, col] - value_func_array[row, col]))

            value_func_array = np.copy(new_value_func_array)

            print(f"\nIteration {iteration}:\n", value_func_array)

            if delta < theta:
                break

        print("\nValue Iteration Converged.")
        return value_func_array
```
